## Tokenization

```python
text = """This is the story of a kingdom torn apart, two powerful kings a homeless foundling, and the black
powder that can destroy them all. The Outcast is a novel almost impossible to describe in a few short sentences. 
At its churning center is Mahalalel, a youth whose very identity was taken from him as an infant. Revolving 
around the mystery of his birth is a fascinating, compelling cast of characters, each with his own history, 
many with their own secrets. There is Ula, the reluctant hero-warrior trying desperately to live up to his adoptive father's
expectations. Azoran, whose insanity is borne of a horrific crime and its aftermath. King Istrak and King
Yosef Diaber, two men so very different who share the same goal: to rule a unified nation.
For five years the kingdom of Khazar has been fractured and darkened, split into two warring sides. 
Now, the fate of the entire nation rests on the thin shoulders of a youth, unwanted and alone.
Compelling, engrossing, and utterly unique, The Outcast takes us into a world we will remember forever."""
```

```python
# import
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
# download module
import nltk
nltk.download("punkt")
```

```python
# sentence tokenization
sent_tokenize(text)
```

```python
# word tokenization
word_tokenize(text)
```

```python
# import regex tokenizer
from nltk.tokenize import RegexpTokenizer
```

```python
# tokenize text - at least one character
tokenizer = RegexpTokenizer(r"\w+")
tokenizer.tokenize(text)
```

```python
# tokenize text - capitalized words
tokenizer = RegexpTokenizer(r"[A-Z]\w+")
tokenizer.tokenize(text)
```

## Removing Stop Words

```python
# imports
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
# tokenize text
words_in_txt = word_tokenize(text)
```

```python
# get english stop words
english_stop_words = stopwords.words("english")
```

```python
# removing stop words from text
filtered_words_in_txt = [word for word in words_in_txt if word.casefold() not in english_stop_words]
filtered_words_in_txt
```

```python
# supported languages
stopwords.fileids()
```

## Stemming

```python
# import
from nltk.stem import PorterStemmer
```

```python
# set stemmer
stemmer = PorterStemmer()
```

```python
text1 = """This is the story of a kingdom torn apart, two powerful kings a homeless foundling, and the black
powder that can destroy them all"""
```

```python
# tokenize text and get stemmed words
words = word_tokenize(text)
stemmed_words = [stemmer.stem(word) for word in words]
stemmed_words
```

```python
# import snowball stemmer
from nltk.stem.snowball import SnowballStemmer
```

```python
# set stemmer
snow_stemmer = SnowballStemmer(language="english")
```

```python
# get stemmed words
stemmed_words = [snow_stemmer.stem(word) for word in words]
stemmed_words
```

```python
# import regex stemmer
from nltk.stem import RegexpStemmer
```

```python
# set stemmer
regex_stemmer = RegexpStemmer("s$|able$|ing$", min=4)
```

```python
# 1
regex_stemmer.stem("available")
```

```python
# 2
regex_stemmer.stem("bless")
```

```python
# 3
regex_stemmer.stem("has")
```

## Lemmatization

```python
# import 
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer
```

```python
# set lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# lemmatize text
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
lemmatized_words
```

```python
# noun
lemmatizer.lemmatize("better")
```

```python
# adjective
lemmatizer.lemmatize("better", pos="a")
```
